I learnt that fastai model only handles jpg files. 

I frist tried to convert every data into a jpg file using opencv but the model traceback. 

I then deleted all files which are not jpg. 

I also deleted some jpg files that cannot be read by hands, but I did not figure out how to do it with code.

```python
cardir = os.path.join(os.getcwd(),'downloads','car')
bikedir = os.path.join(os.getcwd(),'downloads','bike')
```

```python
for car in os.listdir(cardir):
    carImageType = os.path.splitext(car)
    if carImageType[1] == '.jpg':
        None
    else:
        os.remove(os.path.join('downloads','car',car))
```

```python
for bike in os.listdir(bikedir):
    bikeImageType = os.path.splitext(bike)
    if bikeImageType[1] == '.jpg':
        None
    else:
        os.remove(os.path.join('downloads','bike',bike))
```

I seperated datas into train, validation, and test directories with a porpotion of 5:1:1

```python
for car in os.listdir(cardir)[0:int(len(os.listdir(cardir))*5/7)]:
    shutil.copy(os.path.join(cardir,car),os.path.join(os.getcwd(),'train','class1'))
for car in os.listdir(cardir)[int(len(os.listdir(cardir))*5/7):int(len(os.listdir(cardir))*6/7)]:
    shutil.copy(os.path.join(cardir,car),os.path.join(os.getcwd(),'valid','class1'))
for car in os.listdir(cardir)[int(len(os.listdir(cardir))*6/7):]:
    shutil.copy(os.path.join(cardir,car),os.path.join(os.getcwd(),'test','class1'))
for bike in os.listdir(bikedir)[0:int(len(os.listdir(bikedir))*5/7)]:
    shutil.copy(os.path.join(bikedir,bike),os.path.join(os.getcwd(),'train','class2'))
for bike in os.listdir(bikedir)[int(len(os.listdir(bikedir))*5/7):int(len(os.listdir(bikedir))*6/7)]:
    shutil.copy(os.path.join(bikedir,bike),os.path.join(os.getcwd(),'valid','class2'))
for bike in os.listdir(bikedir)[int(len(os.listdir(bikedir))*6/7):]:
    shutil.copy(os.path.join(bikedir,bike),os.path.join(os.getcwd(),'test','class2'))
```
